```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Imports OK")
```

```python
df = pd.read_csv("../data/lake/cardio_deid_data.csv")
print(df.shape)
```

### Plot 1: Population Age Band Distribution (De-identified Cohort)
Shows how the patient cohort is distributed across HIPAA-safe age bands.


```python
plt.figure(figsize=(10,5))
sns.histplot(df["ap_hi"], kde=True, bins=40, color="#4c72b0")
plt.title("Systolic BP Distribution", fontsize=16)
plt.xlabel("Systolic BP")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="bmi_band", palette="Set3")
plt.title("BMI Category Distribution", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["ap_hi"], kde=True, bins=40, color="#4c72b0")
plt.title("Systolic BP Distribution", fontsize=16)
plt.xlabel("Systolic BP")
plt.show()
```

```python
plt.figure(figsize=(8,5))

# Compute counts with correct order
age_order = ["18-29","30-39","40-49","50-59","60-69","70+"]
age_counts = df["age_band"].value_counts().reindex(age_order)

sns.barplot(
    y=age_counts.index,
    x=age_counts.values,
    palette="viridis"
)

plt.title("Age Band Distribution", fontsize=16)
plt.xlabel("Count")
plt.ylabel("Age Band")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("BMI Category Distribution (Clinical Risk Perspective)", fontsize=16)
sns.countplot(data=df, x="bmi_band")
plt.xlabel("BMI Category")
plt.ylabel("Count")
plt.show()
```
